## Tabulating and cross-tabulating data
### cross-tabulation = the construction of frequency tables

```python
# make df
data = {'speaker':["upsy-daisy",  "upsy-daisy",  "upsy-daisy",  "upsy-daisy",  "tombliboo",   "tombliboo",   "makka-pakka", "makka-pakka",
  "makka-pakka", "makka-pakka"],
       'utterance':["pip", "pip", "onk", "onk", "ee",  "oo",  "pip", "pip", "onk", "onk"]}

df = pd.DataFrame(data, columns=['speaker','utterance'])

# cross-tabulate
pd.crosstab(index = df["speaker"], columns = "count")
```

```python
# instead we want to cross-tabulate the speakers with the utterances
pd.crosstab(index=df["speaker"], columns=df["utterance"],margins=True)
```

```python
# organise in proportions instead of raw counts
tabs = pd.crosstab(index=df["speaker"], columns=df["utterance"],margins=True)

tabs.columns = list(tabs.columns)[0:-1] + ['rowtotals']
tabs.index = list(tabs.index)[0:-1] + ['coltotals']

tabs
```

```python
# row-wise proportion i.e. which speaker the utterances come from
tabs.div(tabs["rowtotals"], axis=0)
```

## Transforming or recoding a variable

```python
# in a survey: On a scale of 1 (strongly disagree) to 7 (strongly agree), to what extent do you agree with the proposition that ‚ÄúDinosaurs are awesome‚Äù?
data = [1, 7, 3, 4, 4, 4, 2, 6, 5, 5]
data

# transform likert scale to be from -3 to +3 instead (makes more sense for the survey)
data = [1, 7, 3, 4, 4, 4, 2, 6, 5, 5]
data = [x-4 for x in data]
data
```

```python
# If your data is in a numpy array rather than a list, it is even easier: just subtract 4 from array
import numpy as np
data2 = np.array([1, 7, 3, 4, 4, 4, 2, 6, 5, 5])
data2 = data2 - 4
data2
```

```python
# take the absolute value of the centred data (i.e. strength)
data = np.array([1, 7, 3, 4, 4, 4, 2, 6, 5, 5])
data = data -4
data = abs(data)
data
```

```python
# compute a variable that contains only the direction of the opinion and ignores the strength
data = np.array([1, 7, 3, 4, 4, 4, 2, 6, 5, 5])
data = data - 4
data = np.sign(data) # negative is -1, 0 is 0, and positive is 1
data
```

```python
# collect all this in a dataframe step 1
import pandas as pd
df = pd.DataFrame(
    {'scores': np.array([1, 7, 3, 4, 4, 4, 2, 6, 5, 5])
    })

df
```

```python
# collect all this in a dataframe step 1
df['centered'] = df['scores']-4
df['opinion_strength'] = abs(df['centered'])
df['opinion_direction'] = np.sign(df['scores']-4)
df
```

```python
# cutting a numeric variable up into discrete categoriesüëá

#age = [60,58,24,26,34,42,31,30,33,2,9]
import pandas as pd
df = pd.DataFrame(
    {'age': np.array([60,58,24,26,34,42,31,30,33,2,9])
    })

df
```

```python
# categorize
df['categories'] = pd.cut(x = df['age'], bins = [0,20,40,60], labels = ['young', 'adult', 'older'])
df
```

```python
# divide the age variable into three categories of different size, but with approximately identical numbers of people. This could be done by making the breaks correspond to the 0th, 33rd, 66th and 100th percentiles of the data
df['categories'] = pd.qcut(x = df['age'], q = [0, .33, .66, 1] ) # use qcut() function
df
```

## Slicing and dicing

### Slicing: get chunks of data out of a list. Syntax is as follows: listname[first:last:step_size]. If you put nothing in this third argument, it defaults to one.

```python
# make list
age = [17, 19, 21, 37, 18, 19, 47, 18, 19]

# slice in different ways
age[1:-2]
age[3:6:2]
age[5:len(age)]
```

```python
# "pop" an item out of a list aka remove an item
fruits = ['apples', 'pears', 'bananas', 'spinach', 'strawberries', 'grapes']
fruits.pop(3)
fruits
```

## Extracting a subset of a dataframe

```python
# make dataframe
age = [17, 19, 21, 37, 18, 19, 47, 18, 19]
score = [12, 10, 11, 15, 16, 14, 25, 21, 29]
rt = [3.552, 1.624, 6.431, 7.132, 2.925, 4.662, 3.634, 3.635, 5.234]
group = ["test", "test", "test", "test", "test", "control", "control", "control", "control"]

import pandas as pd

df = pd.DataFrame(
    {'age': age,
     'score': score,
     'rt': rt,
     'group': group
    })

df
```

```python
# only keep data from test group
df_test = df[df['group'] == 'test']
df_test
```

```python
# only keep data from control group
df_control = df[df['group'] == 'control']
df_control
```

```python
# only keep data from subjects over age 21
df_old = df[df['age']> 21] 
df_old
```

```python
# only keep data from subjects between age 17 and 21
df_youngish = df[(df['age'] < 21 ) & (df['age'] > 17)]
df_youngish
```

```python
# only keep data from subject over age 21 and with reaction time over 3 seconds (i.e. slice from more variables at once)
old_and_slow = df[(df['age'] > 21) & (df['rt'] > 3)]
old_and_slow
```

```python
# specify further to only keep data from the control subject over age 21 and with reaction time over 3 seconds
old_and_slow_control = df[(df['age'] > 21) & (df['rt'] > 3) & (df['group'] == 'control')]
old_and_slow_control
```

## Sorting, flipping, or merging datasets

```python
# make dataframe
import pandas as pd


age = [17, 19, 21, 37, 18, 19, 47, 18, 19]
score = [12, 10, 11, 15, 16, 14, 25, 21, 29]
rt = [3.552, 1.624, 6.431, 7.132, 2.925, 4.662, 3.634, 3.635, 5.234]
group = ["test", "test", "test", "test", "test", "control", "control", "control", "control"]

df = pd.DataFrame(
    {'age': age,
     'score': score,
     'rt': rt,
     'group': group
    })

df
```

```python
# sort by column
df_sorted = df.sort_values(by=['age'])
df_sorted
```

```python
# sort from largest to smallest
df_sorted = df.sort_values(by=['age'], ascending = False)
df_sorted
```

```python
# sort by multiple columns
df_sorted = df.sort_values(by=['age', 'score'])
df_sorted
```

```python
# flipping/transposing a dataframe
df_flipped = df.transpose()
df_flipped
```

### Merging dataframes
#### Merge dataframes by adding columns to existing columnsüëá

```python
# create first dataframe
age = [17, 19, 21, 37, 18, 19, 47, 18, 19]
score = [12, 10, 11, 15, 16, 14, 25, 21, 29]
rt = [3.552, 1.624, 6.431, 7.132, 2.925, 4.662, 3.634, 3.635, 5.234]
group = ["test", "test", "test", "test", "test", "control", "control", "control", "control"]

df1 = pd.DataFrame(
    {'age': age,
     'score': score,
     'rt': rt,
     'group': group
    })

df1
```

```python
# merge dataframes - OBS here we merge dataframes to add more variables (columns), not columns
df_joined = df1.join(df2)
df_joined
```

## Reshaping a dataset

### Reshaping the data means taking the data in whatever format it‚Äôs given to you, and converting it to the format you need it


```python
# reshaping from wide to long format

# first, create a wide-format dataframe where each subject corresponds to a row
age = [17, 19, 21, 37, 18, 19, 47, 18, 19]
score = [12, 10, 11, 15, 16, 14, 25, 21, 29]
rt = [3.552, 1.624, 6.431, 7.132, 2.925, 4.662, 3.634, 3.635, 5.234]
group = ["test", "test", "test", "test", "test", "control", "control", "control", "control"]

df = pd.DataFrame(
    {'age': age,
     'score': score,
     'rt': rt,
     'group': group
    })

df
```

```python
# change from wide to long format
df_long = pd.melt(df, id_vars=[ 'age', 'group'])
df_long.head(13)
```
